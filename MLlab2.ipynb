```python

array = np.arange(1, 13)
reshaped_arr = array.reshape(3, 4)

print("Original Array:", array)
print("Original Shape:", array.shape)

print("Reshaped Array:", reshaped_arr)
print("Reshaped Shape:", reshaped_arr.shape)
```
